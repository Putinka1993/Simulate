```python
import pandas as pd
import hashlib

# Предполагаем, что ваш df уже загружен и называется applications
# Если нет - раскомментируйте следующую строку:
# applications = pd.read_csv('ваш_файл.csv', encoding='utf-8')

# Очищаем названия колонок от лишних пробелов
applications.columns = applications.columns.str.strip()

# --- 1. Анализ уникальных значений ДО анонимизации ---
print("="*60)
print("АНАЛИЗ ДО АНОНИМИЗАЦИИ")
print("="*60)

# Список колонок с именами
name_columns = ['Получатель услуг', 'Инициатор', 'Кем решен (сотрудник)']

# Словарь для хранения статистики
stats_before = {}

for col in name_columns:
    if col in applications.columns:
        # Убираем пустые значения и считаем уникальные
        unique_values = applications[col].dropna().unique()
        num_unique = len(unique_values)
        stats_before[col] = {
            'count': num_unique,
            'sample': unique_values[:5]  # Покажем первые 5 для примера
        }
        
        print(f"\nКолонка: {col}")
        print(f"  Уникальных значений: {num_unique}")
        print(f"  Примеры (первые 5): {unique_values[:5]}")
        print(f"  Всего записей в колонке: {applications[col].count()}")
    else:
        print(f"\nКолонка {col} не найдена!")

# Общее количество уникальных людей (объединение всех трех колонок)
all_unique_people = pd.unique(applications[name_columns].values.ravel())
all_unique_people = [p for p in all_unique_people if pd.notna(p)]  # Убираем NaN
print(f"\n{'='*60}")
print(f"ВСЕГО уникальных людей во всех колонках: {len(all_unique_people)}")
print(f"Примеры: {all_unique_people[:10]}")
```

```python
# --- 2. Создаем маппинг "Реальное имя -> ID" ---
print("\n" + "="*60)
print("СОЗДАНИЕ МАППИНГА")
print("="*60)

# Собираем все уникальные имена из всех трех колонок
all_names = []
for col in name_columns:
    all_names.extend(applications[col].dropna().unique())

# Оставляем только уникальные имена
unique_names = list(set(all_names))
unique_names = [name for name in unique_names if pd.notna(name)]  # Убираем NaN

print(f"Всего уникальных имен для маппинга: {len(unique_names)}")

# Создаем словарь имя -> ID
name_to_id = {}
for name in unique_names:
    if pd.notna(name) and name != '':
        # Создаем хеш от имени
        hash_object = hashlib.md5(str(name).encode('utf-8'))
        hash_hex = hash_object.hexdigest()[:8]
        user_id = f"User_{hash_hex}"
        name_to_id[name] = user_id

# Покажем примеры маппинга
print("\nПримеры маппинга (первые 10):")
for i, (name, user_id) in enumerate(list(name_to_id.items())[:10]):
    print(f"  {name} -> {user_id}")

# Проверяем, что все имена попали в словарь
print(f"\nРазмер словаря: {len(name_to_id)} имен")
print(f"Уникальных имен для маппинга: {len(unique_names)}")
print(f"Совпадение: {len(name_to_id) == len(unique_names)}")
```

```python
# --- 3. Применяем анонимизацию ---
print("\n" + "="*60)
print("ПРИМЕНЕНИЕ АНОНИМИЗАЦИИ")
print("="*60)

# Создаем копию, чтобы не менять исходные данные
applications_anon = applications.copy()

# Применяем маппинг к каждой колонке
for col in name_columns:
    if col in applications_anon.columns:
        # Заменяем имена на ID, оставляя NaN без изменений
        applications_anon[col] = applications_anon[col].map(name_to_id).fillna(applications_anon[col])
        
        # Проверяем, что замена произошла
        unique_after = applications_anon[col].dropna().unique()
        print(f"\nКолонка {col}:")
        print(f"  Уникальных значений после замены: {len(unique_after)}")
        print(f"  Примеры ID: {unique_after[:5]}")
```

```python
# --- 4. Финальная проверка ---
print("\n" + "="*60)
print("ФИНАЛЬНАЯ ПРОВЕРКА")
print("="*60)

# Проверяем каждую колонку
print("\nПОКОЛОННОЕ СРАВНЕНИЕ:")
print("-" * 40)

for col in name_columns:
    # Уникальные ДО (без NaN)
    unique_before = set(applications[col].dropna().unique())
    # Уникальные ПОСЛЕ (без NaN)
    unique_after = set(applications_anon[col].dropna().unique())
    
    print(f"\n{col}:")
    print(f"  Уникальных ДО: {len(unique_before)}")
    print(f"  Уникальных ПОСЛЕ: {len(unique_after)}")
    print(f"  Совпадает ли количество? {'ДА' if len(unique_before) == len(unique_after) else 'НЕТ'}")
    
    # Дополнительная проверка: если есть несовпадение, покажем разницу
    if len(unique_before) != len(unique_after):
        print(f"  ВНИМАНИЕ: Несовпадение количества!")

# Проверяем общее количество уникальных людей во всех колонках
all_unique_before = set()
all_unique_after = set()

for col in name_columns:
    all_unique_before.update(applications[col].dropna().unique())
    all_unique_after.update(applications_anon[col].dropna().unique())

print(f"\n{'='*60}")
print(f"ИТОГО УНИКАЛЬНЫХ ЛЮДЕЙ (во всех колонках):")
print(f"  ДО анонимизации: {len(all_unique_before)}")
print(f"  ПОСЛЕ анонимизации: {len(all_unique_after)}")
print(f"  Результат: {'✓ ВСЕ СОХРАНЕНО' if len(all_unique_before) == len(all_unique_after) else '✗ ЕСТЬ ПОТЕРИ'}")
```

```python
# ============================================
# Шаг 3. АНОНИМИЗАЦИЯ УСЛУГ (ПРАВИЛЬНАЯ КАТЕГОРИЗАЦИЯ)
# ============================================

print("\n" + "="*60)
print("ШАГ 3. АНОНИМИЗАЦИЯ УСЛУГ (С КАТЕГОРИЯМИ)")
print("="*60)

# --- 3.1 Оригинальные услуги ---
original_services = applications['Услуга'].dropna().unique()
print(f"\nОригинальных услуг: {len(original_services)}")

# --- 3.2 ПРАВИЛЬНЫЙ маппинг по категориям ---
service_mapping = {
    # РАБОЧИЕ СТАНЦИИ И АРМ (должны стать рабочими станциями)
    'ИТ-АРМ': 'Рабочая станция (стандарт)',
    'ИТ-VIP-пользователь': 'Рабочая станция (повышенной защищенности)',
    'ИТ-АРМ с удаленным доступом, VDI': 'Тонкий клиент / VDI',
    '1-я линия поддержки': 'Рабочая станция оператора',
    'ИТ-Система КЦ': 'Рабочая станция (КЦ)',  # Коллективный центр - тоже рабочая станция
    
    # СЕТЕВОЕ ОБОРУДОВАНИЕ
    'ИТ-СКС': 'Сетевое оборудование (пассивное)',
    'ИТ-Оборудование ЛВС': 'Сетевое оборудование (активное)',
    'ИТ-Интернет': 'Доступ в интернет (сетевое)',
    
    # СЕРВЕРЫ
    'ИТ-Поддержка серверов (программная)': 'Серверное ПО',
    'ИТ-Поддержка серверов в части программного обеспечения': 'Серверная инфраструктура',
    'ИТ-Мониторинг': 'Система мониторинга серверов',
    
    # ОРГТЕХНИКА
    'ИТ-Оргтехника': 'Оргтехника (принтеры/МФУ)',
    
    # СИСТЕМЫ БЕЗОПАСНОСТИ
    'ИТ-Поддержка процессов информационной безопасности': 'СЗИ (антивирусы/шифрование)',
    'ИТ-Поддержка систем информационной безопасности': 'Специализированные СЗИ',
    'ИТ-Поддержка СИРИУС-СБ': 'Система безопасности СИРИУС',
    
    # СПЕЦИАЛИЗИРОВАННЫЕ СИСТЕМЫ
    'ИТ-Система отображения информации': 'Система отображения (табло/мониторы)',
    'ИТ-Система видеонаблюдения': 'Система видеонаблюдения',
    'ИТ-Система электроснабжения': 'СКУД и электроснабжение',
    'ИТ-Система часофикации': 'Система единого времени',
    'ИТ-Система управления инженерными данными': 'Инженерные САПР',
    'ИТ-Система управления строительством': 'Строительные информационные системы',
    'ИТ-Система управления сбытом': 'Сбытовые информационные системы',
    'ИТ-Аналитическая система общего назначения': 'BI-системы',
    'ИТ-Диспетчерская служба': 'Диспетчерские системы',
    
    # ПОЧТА И ДОКУМЕНТЫ
    'ИТ-Электронная почта': 'Корпоративная почта',
    'ИТ-Документооборот': 'СЭД',
    
    # УЧЕТНЫЕ ЗАПИСИ
    'ИТ-Учетная запись домена': 'Учетные записи (AD)',
    'ИТ-Доступа к файловым сетевым ресурсам': 'Файловые ресурсы и доступы',
    
    # СВЯЗЬ
    'ИТ-ВКС': 'Видео-конференц-связь',
    'ИТ-Телефония': 'IP-телефония',
    'ИТ-Системы связи и оповещений': 'Системы оповещения',
    
    # ОСТАЛЬНОЕ (подберем категории)
    'ИТ-Порталы и сайты': 'Корпоративный портал',
    'ИТ-Корпоративные справочники НСИ': 'НСИ и справочники',
    'ИТ-Сбор и обработка отчетности': 'Отчетные системы',
    'ИТ-DigitalAtom': 'Цифровая платформа',
    'ИТ-ИТ-актив': 'Учет ИТ-активов',
    'ИТ-Система клиент-банк': 'Финансовые системы',
    
    # ВНЕШНИЕ УСЛУГИ
    'ВН-1-2 Внешние услуги': 'Внешние ИТ-услуги',
    'ВН-3-1-Услуги ОИКТ': 'Услуги ОИКТ',
    'ПР-0-0-Проекты': 'Проектная деятельность',
    'БК - Информационные системы': 'Банковские информационные системы',
    'КОС-7-1-Внутренние работы': 'Внутренние работы'
}

# Проверяем, все ли оригинальные услуги попали в маппинг
missing = [s for s in original_services if s not in service_mapping]
if missing:
    print(f"\nПропущены услуги: {missing}")
    # Добавляем пропущенные
    for s in missing:
        service_mapping[s] = f"ИТ-услуга ({s.split('-')[-1]})"

print(f"\nМаппинг создан для {len(service_mapping)} услуг")

# Группируем по категориям для проверки
categories = {}
for orig, synth in service_mapping.items():
    if 'Рабочая станция' in synth:
        cat = 'Рабочие станции'
    elif 'Сетев' in synth:
        cat = 'Сети'
    elif 'Сервер' in synth:
        cat = 'Серверы'
    elif 'Оргтехника' in synth:
        cat = 'Оргтехника'
    elif 'СЗИ' in synth or 'безопасн' in synth:
        cat = 'Безопасность'
    elif 'видео' in synth.lower() or 'отображ' in synth.lower():
        cat = 'Спецсистемы'
    elif 'почт' in synth.lower() or 'сэд' in synth.lower():
        cat = 'Документооборот'
    else:
        cat = 'Прочее'
    
    if cat not in categories:
        categories[cat] = []
    categories[cat].append((orig, synth))

print("\nРаспределение по категориям:")
for cat, items in categories.items():
    print(f"  {cat}: {len(items)} услуг")
    # Покажем примеры
    for orig, synth in items[:2]:
        print(f"    {orig} -> {synth}")
    if len(items) > 2:
        print(f"    ...")

# Применяем маппинг
service_col = None
for col in applications_anon.columns:
    if 'Услуга' in col:
        service_col = col
        break

if service_col:
    applications_anon[service_col] = applications[service_col].map(service_mapping).fillna(applications[service_col])
    print(f"\n✓ Замена применена")
    
    # Покажем результат
    print("\nИтоговые услуги (первые 15):")
    for service in sorted(applications_anon[service_col].unique())[:15]:
        print(f"  {service}")
```

```python
# ============================================
# Шаг 4. ГЕНЕРАЦИЯ ОПИСАНИЙ ДЛЯ ВСЕХ КАТЕГОРИЙ
# ============================================
import random

print("\n" + "="*60)
print("ШАГ 4. ГЕНЕРАЦИЯ ОПИСАНИЙ (ВСЕ КАТЕГОРИИ)")
print("="*60)

# --- Определяем категории услуг по их названиям ---
def get_detailed_category(service_name):
    """Точное определение категории услуги"""
    s = str(service_name).lower()
    
    if 'рабочая станция' in s or 'арм' in s or 'тонкий клиент' in s or 'vdi' in s:
        return 'workstation'
    elif 'сетев' in s or 'лвс' in s or 'скс' in s or 'доступ в интернет' in s:
        return 'network'
    elif 'сервер' in s or 'мониторинг' in s:
        return 'server'
    elif 'оргтехника' in s or 'принтер' in s or 'мфу' in s:
        return 'printers'
    elif 'сзи' in s or 'безопасн' in s or 'антивирус' in s or 'сириус' in s:
        return 'security'
    elif 'видеонаблюд' in s or 'отображен' in s or 'часофикац' in s or 'электроснабж' in s:
        return 'special'
    elif 'почт' in s or 'сэд' in s or 'документооборот' in s:
        return 'docflow'
    elif 'вкс' in s or 'телефон' in s or 'связи' in s:
        return 'communication'
    elif 'учетн' in s or 'домен' in s or 'файлов' in s:
        return 'access'
    else:
        return 'other'

# --- ШАБЛОНЫ ДЛЯ КАЖДОЙ КАТЕГОРИИ ---
templates_by_category = {
    'workstation': [
        "На рабочей станции {problem}. Требуется помощь.",
        "Не загружается {os} на ПК пользователя.",
        "После обновления {software} перестал работать {function}.",
        "Прошу установить {software} на компьютер в {location}.",
        "Синий экран при запуске. Ошибка: {error_code}",
        "Зависает {software} при работе с документами."
    ],
    
    'network': [
        "Пропал интернет в {location}. Не пингуется {resource}.",
        "Не работает {equipment} в {location}. Нет связи с серверами.",
        "Требуется настроить доступ к {resource} из {location}.",
        "Медленно работает сеть в {location}. Скорость упала.",
        "Не подключается к Wi-Fi в {location}. Точка доступа не видна."
    ],
    
    'server': [
        "Не отвечает сервер {server_name}. Ошибка подключения.",
        "Высокая нагрузка на CPU сервера {server_name}.",
        "Сбой службы на сервере. Требуется перезапуск.",
        "Закончилось место на диске сервера {server_name}.",
        "Ошибка резервного копирования на сервере {server_name}."
    ],
    
    'printers': [
        "Не печатает {printer_model} в {location}. Ошибка подачи бумаги.",
        "Закончился тонер в принтере. Требуется замена картриджа.",
        "Застряла бумага в МФУ. Нужна помощь.",
        "Сканер не видит документы. Ошибка подключения.",
        "Принтер выдает ошибку {error_code}. Требуется диагностика."
    ],
    
    'security': [
        "Антивирус заблокировал {software}. Требуется исключение.",
        "Сработала сигнализация СЗИ на рабочей станции.",
        "Не открывается {resource}. Требуются права доступа.",
        "Истек срок действия сертификата у пользователя.",
        "Обнаружена подозрительная активность в системе."
    ],
    
    'special': [
        "Не работает камера {camera_id} в {location}. Нет сигнала.",
        "Сбой в системе отображения информации. Не обновляются данные.",
        "Рассинхронизация времени на оборудовании.",
        "Нет данных с датчиков в {location}. Требуется проверка.",
        "Ошибка отображения на табло коллективного пользования."
    ],
    
    'docflow': [
        "Не отправляется письмо с ошибкой {error_code}.",
        "Не приходит почта на ящик пользователя.",
        "Зависает СЭД при открытии документа.",
        "Требуется настроить шаблон документа в СЭД.",
        "Ошибка согласования документа в системе."
    ],
    
    'communication': [
        "Нет звука в {location} при ВКС.",
        "Плохая слышимость по телефону. Шумы.",
        "Не работает микрофон в переговорной {location}.",
        "Сбрасывается соединение при ВКС.",
        "Телефон не регистрируется в сети. Ошибка авторизации."
    ],
    
    'access': [
        "Не могу зайти в {system} под учетной записью.",
        "Требуется создать учетную запись для нового сотрудника.",
        "Нет доступа к папке {folder} на сервере.",
        "Истек срок действия пароля. Прошу разблокировать.",
        "Учетная запись заблокирована после неудачных попыток."
    ],
    
    'other': [
        "Требуется обслуживание по заявке.",
        "Заявка на выполнение работ.",
        "Прошу помочь с настройкой."
    ]
}

# --- Словари для подстановки ---
values = {
    'problem': ['зависает', 'тормозит', 'выдает ошибку', 'не включается', 'перезагружается'],
    'os': ['Windows 10', 'Windows 11', 'Astra Linux', 'ALT Linux', 'РЕД ОС'],
    'software': ['Microsoft Office', 'LibreOffice', '1С:Предприятие', 'САПР', 'СЭД'],
    'function': ['печать', 'сохранение', 'отправка почты', 'экспорт данных', 'открытие документов'],
    'error_code': ['0x80070005', '0x80004005', '0x80070002', '0x80070570'],
    'equipment': ['коммутатор', 'маршрутизатор', 'точка доступа', 'медиаконвертер'],
    'resource': ['сервер', 'базу данных', 'корпоративный портал', '1С', 'файловый ресурс'],
    'server_name': ['SRV-DC01', 'SRV-FS02', 'SRV-SQL01', 'SRV-APP01', 'SRV-BACKUP'],
    'service': ['служба DNS', 'служба DHCP', 'Active Directory', 'служба печати'],
    'printer_model': ['HP LaserJet', 'Xerox', 'Kyocera', 'Canon', 'Brother'],
    'camera_id': [f'CAM-{i:03d}' for i in range(1, 11)],
    'city': ['Москва', 'Санкт-Петербург', 'Воронеж', 'Саратов'],
    'phone': ['101-12', '102-34', '103-56', '104-78'],
    'folder': ['Отчеты', 'Документация', 'Проекты', 'Сметы', 'Общая'],
    'system': ['Active Directory', '1С', 'СЭД', 'почтовую систему', 'корпоративный портал']
}

# --- ОПРЕДЕЛЯЕМ ВСЕ НУЖНЫЕ КОЛОНКИ ---
print("\nОпределяем колонки...")

service_col = None
request_col = None
description_col = None
location_col = None

for col in applications_anon.columns:
    if 'Услуга' in col:
        service_col = col
        print(f"  Найдена колонка услуги: '{service_col}'")
    elif 'Вид запроса' in col:
        request_col = col
        print(f"  Найдена колонка вида запроса: '{request_col}'")
    elif 'Описание' in col:
        description_col = col
        print(f"  Найдена колонка описания: '{description_col}'")
    elif 'Расположение' in col:
        location_col = col
        print(f"  Найдена колонка расположения: '{location_col}'")

# Проверяем, что все нужные колонки найдены
if not service_col:
    raise ValueError("Не найдена колонка с услугой!")
if not request_col:
    raise ValueError("Не найдена колонка с видом запроса!")
if not description_col:
    raise ValueError("Не найдена колонка с описанием!")

print("\n✓ Все необходимые колонки найдены")

# --- Генерация описаний ---
print("\nГенерируем описания...")

new_descriptions = []
category_stats = {}

for idx, row in applications_anon.iterrows():
    service = row[service_col]
    req_type = row[request_col]
    location = row.get(location_col, 'офис') if location_col else 'офис'
    
    # Определяем категорию
    category = get_detailed_category(service)
    category_stats[category] = category_stats.get(category, 0) + 1
    
    # Выбираем шаблон
    templates = templates_by_category.get(category, templates_by_category['other'])
    template = random.choice(templates)
    
    # Подготавливаем переменные
    template_vars = {}
    for key in values.keys():
        template_vars[key] = random.choice(values[key])
    template_vars['location'] = location
    template_vars['request_type'] = req_type
    
    # Формируем описание
    try:
        description = template.format(**template_vars)
    except KeyError as e:
        description = f"Заявка по услуге {service}. Тип: {req_type}"
    
    # Добавляем вежливость
    if random.random() > 0.5:
        description = "Добрый день! " + description[0].lower() + description[1:]
    
    new_descriptions.append(description)

# Применяем
applications_anon[description_col] = new_descriptions

# --- Статистика ---
print("\nСтатистика по категориям:")
total = len(applications_anon)
for cat, count in sorted(category_stats.items()):
    print(f"  {cat}: {count} заявок ({count/total*100:.1f}%)")

print(f"\nВсего обработано: {total} заявок")

# Показываем примеры
print("\nПримеры сгенерированных описаний по категориям:")
for category in ['workstation', 'network', 'server', 'printers', 'security', 'special', 'docflow', 'communication', 'access']:
    examples = []
    for idx, row in applications_anon.iterrows():
        if get_detailed_category(row[service_col]) == category and len(examples) < 2:
            examples.append(row[description_col])
    
    if examples:
        print(f"\n{category.upper()}:")
        for ex in examples:
            print(f"  - {ex}")
```
