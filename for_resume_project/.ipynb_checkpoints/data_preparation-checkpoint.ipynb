```python
import pandas as pd
import hashlib

# Предполагаем, что ваш df уже загружен и называется applications
# Если нет - раскомментируйте следующую строку:
# applications = pd.read_csv('ваш_файл.csv', encoding='utf-8')

# Очищаем названия колонок от лишних пробелов
applications.columns = applications.columns.str.strip()

# --- 1. Анализ уникальных значений ДО анонимизации ---
print("="*60)
print("АНАЛИЗ ДО АНОНИМИЗАЦИИ")
print("="*60)

# Список колонок с именами
name_columns = ['Получатель услуг', 'Инициатор', 'Кем решен (сотрудник)']

# Словарь для хранения статистики
stats_before = {}

for col in name_columns:
    if col in applications.columns:
        # Убираем пустые значения и считаем уникальные
        unique_values = applications[col].dropna().unique()
        num_unique = len(unique_values)
        stats_before[col] = {
            'count': num_unique,
            'sample': unique_values[:5]  # Покажем первые 5 для примера
        }
        
        print(f"\nКолонка: {col}")
        print(f"  Уникальных значений: {num_unique}")
        print(f"  Примеры (первые 5): {unique_values[:5]}")
        print(f"  Всего записей в колонке: {applications[col].count()}")
    else:
        print(f"\nКолонка {col} не найдена!")

# Общее количество уникальных людей (объединение всех трех колонок)
all_unique_people = pd.unique(applications[name_columns].values.ravel())
all_unique_people = [p for p in all_unique_people if pd.notna(p)]  # Убираем NaN
print(f"\n{'='*60}")
print(f"ВСЕГО уникальных людей во всех колонках: {len(all_unique_people)}")
print(f"Примеры: {all_unique_people[:10]}")
```

```python
# --- 2. Создаем маппинг "Реальное имя -> ID" ---
print("\n" + "="*60)
print("СОЗДАНИЕ МАППИНГА")
print("="*60)

# Собираем все уникальные имена из всех трех колонок
all_names = []
for col in name_columns:
    all_names.extend(applications[col].dropna().unique())

# Оставляем только уникальные имена
unique_names = list(set(all_names))
unique_names = [name for name in unique_names if pd.notna(name)]  # Убираем NaN

print(f"Всего уникальных имен для маппинга: {len(unique_names)}")

# Создаем словарь имя -> ID
name_to_id = {}
for name in unique_names:
    if pd.notna(name) and name != '':
        # Создаем хеш от имени
        hash_object = hashlib.md5(str(name).encode('utf-8'))
        hash_hex = hash_object.hexdigest()[:8]
        user_id = f"User_{hash_hex}"
        name_to_id[name] = user_id

# Покажем примеры маппинга
print("\nПримеры маппинга (первые 10):")
for i, (name, user_id) in enumerate(list(name_to_id.items())[:10]):
    print(f"  {name} -> {user_id}")

# Проверяем, что все имена попали в словарь
print(f"\nРазмер словаря: {len(name_to_id)} имен")
print(f"Уникальных имен для маппинга: {len(unique_names)}")
print(f"Совпадение: {len(name_to_id) == len(unique_names)}")
```

```python
# --- 3. Применяем анонимизацию ---
print("\n" + "="*60)
print("ПРИМЕНЕНИЕ АНОНИМИЗАЦИИ")
print("="*60)

# Создаем копию, чтобы не менять исходные данные
applications_anon = applications.copy()

# Применяем маппинг к каждой колонке
for col in name_columns:
    if col in applications_anon.columns:
        # Заменяем имена на ID, оставляя NaN без изменений
        applications_anon[col] = applications_anon[col].map(name_to_id).fillna(applications_anon[col])
        
        # Проверяем, что замена произошла
        unique_after = applications_anon[col].dropna().unique()
        print(f"\nКолонка {col}:")
        print(f"  Уникальных значений после замены: {len(unique_after)}")
        print(f"  Примеры ID: {unique_after[:5]}")
```

```python
# --- 4. Финальная проверка ---
print("\n" + "="*60)
print("ФИНАЛЬНАЯ ПРОВЕРКА")
print("="*60)

# Проверяем каждую колонку
print("\nПОКОЛОННОЕ СРАВНЕНИЕ:")
print("-" * 40)

for col in name_columns:
    # Уникальные ДО (без NaN)
    unique_before = set(applications[col].dropna().unique())
    # Уникальные ПОСЛЕ (без NaN)
    unique_after = set(applications_anon[col].dropna().unique())
    
    print(f"\n{col}:")
    print(f"  Уникальных ДО: {len(unique_before)}")
    print(f"  Уникальных ПОСЛЕ: {len(unique_after)}")
    print(f"  Совпадает ли количество? {'ДА' if len(unique_before) == len(unique_after) else 'НЕТ'}")
    
    # Дополнительная проверка: если есть несовпадение, покажем разницу
    if len(unique_before) != len(unique_after):
        print(f"  ВНИМАНИЕ: Несовпадение количества!")

# Проверяем общее количество уникальных людей во всех колонках
all_unique_before = set()
all_unique_after = set()

for col in name_columns:
    all_unique_before.update(applications[col].dropna().unique())
    all_unique_after.update(applications_anon[col].dropna().unique())

print(f"\n{'='*60}")
print(f"ИТОГО УНИКАЛЬНЫХ ЛЮДЕЙ (во всех колонках):")
print(f"  ДО анонимизации: {len(all_unique_before)}")
print(f"  ПОСЛЕ анонимизации: {len(all_unique_after)}")
print(f"  Результат: {'✓ ВСЕ СОХРАНЕНО' if len(all_unique_before) == len(all_unique_after) else '✗ ЕСТЬ ПОТЕРИ'}")
```

```python
# --- 5. Сохраняем анонимизированный датасет и маппинг ---

# Сохраняем основной датасет
applications_anon.to_csv(r'C:\Users\Incognitus\Documents\раб_данные\applications_anonymized.csv', index=False, encoding='utf-8-sig')
print("\n✓ Анонимизированный датасет сохранен: 'applications_anonymized.csv'")

# Сохраняем маппинг на случай, если понадобится расшифровка (но обычно это не нужно)
mapping_df = pd.DataFrame(list(name_to_id.items()), columns=['Original_Name', 'User_ID'])
mapping_df.to_csv('name_mapping.csv', index=False, encoding='utf-8-sig')
print("✓ Маппинг имен сохранен: 'name_mapping.csv' (храните в безопасном месте)")

# Покажем финальный результат
print("\n" + "="*60)
print("ПРИМЕР РЕЗУЛЬТАТА (первые 3 строки):")
print("="*60)
print("\nДО анонимизации:")
print(applications[name_columns].head(3))

print("\nПОСЛЕ анонимизации:")
print(applications_anon[name_columns].head(3))
```

```python
# ============================================
# Шаг 2. Анонимизация РАСПОЛОЖЕНИЯ
# ============================================

print("\n" + "="*60)
print("ШАГ 2. АНОНИМИЗАЦИЯ РАСПОЛОЖЕНИЯ")
print("="*60)

# --- 2.1 Анализируем уникальные значения ДО замены ---
original_locations = applications['Расположение'].dropna().unique()
num_locations = len(original_locations)

print(f"\nДО замены:")
print(f"  Уникальных объектов: {num_locations}")
print(f"  Примеры реальных названий: {original_locations[:5]}")

# --- 2.2 Создаем синтетические названия для атомных станций и объектов ---
# Генерируем 18 вымышленных названий
synthetic_locations = [
    # Атомные станции
    "АЭС-1 (Северная)",
    "АЭС-2 (Западная)",
    "АЭС-3 (Восточная)",
    "АЭС-4 (Южная)",
    "АЭС-5 (Центральная)",
    "АЭС-6 (Прибрежная)",
    # Объекты и площадки
    "Промплощадка Альфа",
    "Промплощадка Бета",
    "Хранилище отработанного топлива №1",
    "Хранилище отработанного топлива №2",
    "Учебно-тренировочный центр",
    "Лабораторный корпус",
    "Административно-бытовой корпус",
    "Здание управления (ЗУ-1)",
    "Здание управления (ЗУ-2)",
    "Резервный пункт управления",
    "Складское хозяйство",
    "Турбинный цех"
]

# Проверяем, что количество совпадает
print(f"\nСгенерировано синтетических названий: {len(synthetic_locations)}")
print(f"  Примеры новых названий: {synthetic_locations[:5]}")

if len(synthetic_locations) != num_locations:
    print(f"  ВНИМАНИЕ: Количество не совпадает! Нужно {num_locations}, создано {len(synthetic_locations)}")
    # Дополняем если нужно
    while len(synthetic_locations) < num_locations:
        synthetic_locations.append(f"Объект Special-{len(synthetic_locations)+1}")
else:
    print(f"  ✓ Количество совпадает: {num_locations}")

# --- 2.3 Создаем маппинг и применяем замену ---
location_mapping = dict(zip(original_locations, synthetic_locations))

print("\nПримеры маппинга (первые 5):")
for i, (orig, synth) in enumerate(list(location_mapping.items())[:5]):
    print(f"  {orig} -> {synth}")

# Применяем замену
applications_anon['Расположение'] = applications['Расположение'].map(location_mapping).fillna(applications['Расположение'])

# --- 2.4 Проверка ---
new_locations = applications_anon['Расположение'].dropna().unique()
print(f"\nПОСЛЕ замены:")
print(f"  Уникальных объектов: {len(new_locations)}")
print(f"  Примеры новых названий: {new_locations[:5]}")
print(f"  Совпадает ли количество? {'✓ ДА' if len(new_locations) == num_locations else '✗ НЕТ'}")
```

```python
# ============================================
# Шаг 3. АНОНИМИЗАЦИЯ УСЛУГ (ПРАВИЛЬНАЯ КАТЕГОРИЗАЦИЯ)
# ============================================

print("\n" + "="*60)
print("ШАГ 3. АНОНИМИЗАЦИЯ УСЛУГ (С КАТЕГОРИЯМИ)")
print("="*60)

# --- 3.1 Оригинальные услуги ---
original_services = applications['Услуга'].dropna().unique()
print(f"\nОригинальных услуг: {len(original_services)}")

# --- 3.2 ПРАВИЛЬНЫЙ маппинг по категориям ---
service_mapping = {
    # РАБОЧИЕ СТАНЦИИ И АРМ (должны стать рабочими станциями)
    'ИТ-АРМ': 'Рабочая станция (стандарт)',
    'ИТ-VIP-пользователь': 'Рабочая станция (повышенной защищенности)',
    'ИТ-АРМ с удаленным доступом, VDI': 'Тонкий клиент / VDI',
    '1-я линия поддержки': 'Рабочая станция оператора',
    'ИТ-Система КЦ': 'Рабочая станция (КЦ)',  # Коллективный центр - тоже рабочая станция
    
    # СЕТЕВОЕ ОБОРУДОВАНИЕ
    'ИТ-СКС': 'Сетевое оборудование (пассивное)',
    'ИТ-Оборудование ЛВС': 'Сетевое оборудование (активное)',
    'ИТ-Интернет': 'Доступ в интернет (сетевое)',
    
    # СЕРВЕРЫ
    'ИТ-Поддержка серверов (программная)': 'Серверное ПО',
    'ИТ-Поддержка серверов в части программного обеспечения': 'Серверная инфраструктура',
    'ИТ-Мониторинг': 'Система мониторинга серверов',
    
    # ОРГТЕХНИКА
    'ИТ-Оргтехника': 'Оргтехника (принтеры/МФУ)',
    
    # СИСТЕМЫ БЕЗОПАСНОСТИ
    'ИТ-Поддержка процессов информационной безопасности': 'СЗИ (антивирусы/шифрование)',
    'ИТ-Поддержка систем информационной безопасности': 'Специализированные СЗИ',
    'ИТ-Поддержка СИРИУС-СБ': 'Система безопасности СИРИУС',
    
    # СПЕЦИАЛИЗИРОВАННЫЕ СИСТЕМЫ
    'ИТ-Система отображения информации': 'Система отображения (табло/мониторы)',
    'ИТ-Система видеонаблюдения': 'Система видеонаблюдения',
    'ИТ-Система электроснабжения': 'СКУД и электроснабжение',
    'ИТ-Система часофикации': 'Система единого времени',
    'ИТ-Система управления инженерными данными': 'Инженерные САПР',
    'ИТ-Система управления строительством': 'Строительные информационные системы',
    'ИТ-Система управления сбытом': 'Сбытовые информационные системы',
    'ИТ-Аналитическая система общего назначения': 'BI-системы',
    'ИТ-Диспетчерская служба': 'Диспетчерские системы',
    
    # ПОЧТА И ДОКУМЕНТЫ
    'ИТ-Электронная почта': 'Корпоративная почта',
    'ИТ-Документооборот': 'СЭД',
    
    # УЧЕТНЫЕ ЗАПИСИ
    'ИТ-Учетная запись домена': 'Учетные записи (AD)',
    'ИТ-Доступа к файловым сетевым ресурсам': 'Файловые ресурсы и доступы',
    
    # СВЯЗЬ
    'ИТ-ВКС': 'Видео-конференц-связь',
    'ИТ-Телефония': 'IP-телефония',
    'ИТ-Системы связи и оповещений': 'Системы оповещения',
    
    # ОСТАЛЬНОЕ (подберем категории)
    'ИТ-Порталы и сайты': 'Корпоративный портал',
    'ИТ-Корпоративные справочники НСИ': 'НСИ и справочники',
    'ИТ-Сбор и обработка отчетности': 'Отчетные системы',
    'ИТ-DigitalAtom': 'Цифровая платформа',
    'ИТ-ИТ-актив': 'Учет ИТ-активов',
    'ИТ-Система клиент-банк': 'Финансовые системы',
    
    # ВНЕШНИЕ УСЛУГИ
    'ВН-1-2 Внешние услуги': 'Внешние ИТ-услуги',
    'ВН-3-1-Услуги ОИКТ': 'Услуги ОИКТ',
    'ПР-0-0-Проекты': 'Проектная деятельность',
    'БК - Информационные системы': 'Банковские информационные системы',
    'КОС-7-1-Внутренние работы': 'Внутренние работы'
}

# Проверяем, все ли оригинальные услуги попали в маппинг
missing = [s for s in original_services if s not in service_mapping]
if missing:
    print(f"\nПропущены услуги: {missing}")
    # Добавляем пропущенные
    for s in missing:
        service_mapping[s] = f"ИТ-услуга ({s.split('-')[-1]})"

print(f"\nМаппинг создан для {len(service_mapping)} услуг")

# Группируем по категориям для проверки
categories = {}
for orig, synth in service_mapping.items():
    if 'Рабочая станция' in synth:
        cat = 'Рабочие станции'
    elif 'Сетев' in synth:
        cat = 'Сети'
    elif 'Сервер' in synth:
        cat = 'Серверы'
    elif 'Оргтехника' in synth:
        cat = 'Оргтехника'
    elif 'СЗИ' in synth or 'безопасн' in synth:
        cat = 'Безопасность'
    elif 'видео' in synth.lower() or 'отображ' in synth.lower():
        cat = 'Спецсистемы'
    elif 'почт' in synth.lower() or 'сэд' in synth.lower():
        cat = 'Документооборот'
    else:
        cat = 'Прочее'
    
    if cat not in categories:
        categories[cat] = []
    categories[cat].append((orig, synth))

print("\nРаспределение по категориям:")
for cat, items in categories.items():
    print(f"  {cat}: {len(items)} услуг")
    # Покажем примеры
    for orig, synth in items[:2]:
        print(f"    {orig} -> {synth}")
    if len(items) > 2:
        print(f"    ...")

# Применяем маппинг
service_col = None
for col in applications_anon.columns:
    if 'Услуга' in col:
        service_col = col
        break

if service_col:
    applications_anon[service_col] = applications[service_col].map(service_mapping).fillna(applications[service_col])
    print(f"\n✓ Замена применена")
    
    # Покажем результат
    print("\nИтоговые услуги (первые 15):")
    for service in sorted(applications_anon[service_col].unique())[:15]:
        print(f"  {service}")
```

```python
# ============================================
# Шаг 4. ГЕНЕРАЦИЯ СИНТЕТИЧЕСКИХ ОПИСАНИЙ (ОСМЫСЛЕННАЯ)
# ============================================

print("\n" + "="*60)
print("ШАГ 4. ГЕНЕРАЦИЯ СИНТЕТИЧЕСКИХ ОПИСАНИЙ (ОСМЫСЛЕННАЯ)")
print("="*60)

# --- 4.1 Создаем расширенные шаблоны для разных категорий услуг ---

# Определяем категории услуг на основе наших новых названий
service_category_templates = {
    # 1. РАБОЧИЕ СТАНЦИИ
    'Рабочая станция': {
        'установка': [
            "Требуется установить {software} на рабочую станцию в {location}. Пользователь: {user}.",
            "Прошу установить {software} на новый ПК для сотрудника {user}.",
            "Необходимо установить {software} для выполнения задач в {system}."
        ],
        'настройка': [
            "На рабочей станции {problem}. Требуется настройка {software}.",
            "После обновления {software} перестал работать {function}. Настройте, пожалуйста.",
            "Некорректно работает {software}. Выполните настройку."
        ],
        'не работает': [
            "Не включается рабочая станция в {location}. Требуется диагностика.",
            "Синий экран при загрузке рабочей станции пользователя {user}.",
            "Рабочая станция зависает при работе в {software}."
        ],
        'доступ': [
            "Нет доступа к {resource} с рабочей станции пользователя {user}.",
            "Требуется открыть доступ к {resource} для рабочей станции в {location}."
        ],
        'консультация': [
            "Подскажите, как настроить {software} на рабочей станции для {task}.",
            "Нужна консультация по работе в {software} на новой рабочей станции."
        ]
    },
    
    # 2. АРМ РУКОВОДИТЕЛЯ
    'АРМ руководителя': {
        'установка': [
            "Требуется установка {software} на АРМ руководителя {user}.",
            "Установите {software} для руководителя {user} в срочном порядке."
        ],
        'настройка': [
            "Настройте АРМ руководителя для работы с {system}. Требуется доступ к {resource}.",
            "Перенастройте рабочее место руководителя {user} после переезда в кабинет {location}."
        ],
        'не работает': [
            "Не загружается АРМ руководителя {user}. Срочно требуется помощь.",
            "Проблемы с отображением {software} на АРМ руководителя."
        ]
    },
    
    # 3. СЕТЕВОЕ ОБОРУДОВАНИЕ
    'Сетевое оборудование': {
        'не работает': [
            "Пропала сеть в {location}. Не работает {equipment}.",
            "Отключился {equipment} в {location}. Пользователи сидят без сети.",
            "Мигает {equipment}, пропадает связь с серверами."
        ],
        'настройка': [
            "Требуется настроить {equipment} для нового сегмента сети в {location}.",
            "Выполните настройку VLAN на {equipment} согласно документации.",
            "Необходимо изменить конфигурацию {equipment}."
        ],
        'консультация': [
            "Подскажите, как подключить новое {equipment} к существующей сети."
        ]
    },
    
    # 4. СЕРВЕРЫ
    'Серверная инфраструктура': {
        'не работает': [
            "Не отвечает сервер {server_name}. Ошибка: {error}.",
            "Высокая нагрузка на сервер {server_name}, пользователи жалуются на тормоза.",
            "Сбой службы {service} на сервере {server_name}."
        ],
        'настройка': [
            "Требуется настроить {service} на сервере {server_name}.",
            "Выполните настройку резервного копирования для сервера {server_name}."
        ],
        'обновление': [
            "Необходимо установить обновления безопасности на сервер {server_name}.",
            "Проведите обновление {software} на сервере {server_name}."
        ]
    },
    
    # 5. ПРИНТЕРЫ И ОРГТЕХНИКА
    'Оргтехника': {
        'не работает': [
            "Не печатает принтер в {location}. Ошибка: {error}.",
            "Сканер в {location} не видит документы. Требуется диагностика.",
            "Застряла бумага в МФУ в {location}. Нужна помощь."
        ],
        'заправка': [
            "Закончился тонер в принтере {location}. Требуется замена картриджа.",
            "Прошу заправить картридж для принтера в {location}."
        ],
        'установка': [
            "Требуется установить драйверы для нового МФУ в {location}.",
            "Подключите и настройте принтер для пользователей в {location}."
        ]
    },
    
    # 6. ПОЧТА И ДОКУМЕНТООБОРОТ
    'Корпоративная почта': {
        'не работает': [
            "Не отправляются письма с почты пользователя {user}.",
            "Не приходит почта на ящик пользователя {user}.",
            "Ошибка аутентификации в почте у пользователя {user}."
        ],
        'доступ': [
            "Создайте почтовый ящик для нового сотрудника {user}.",
            "Заблокируйте почту уволенного сотрудника {user}.",
            "Восстановите доступ к почте пользователя {user}."
        ],
        'настройка': [
            "Настройте почтовый клиент на рабочей станции пользователя {user}.",
            "Перенаправьте почту с ящика {user} на {other_user}."
        ]
    },
    
    # 7. СИСТЕМЫ БЕЗОПАСНОСТИ
    'Средства защиты информации': {
        'проблема': [
            "Антивирус блокирует работу {software} у пользователя {user}.",
            "Сработала сигнализация СЗИ на рабочей станции {user}.",
            "Требуется исключение в антивирусе для {software}."
        ],
        'доступ': [
            "Выдайте сертификат для доступа к {system} пользователю {user}.",
            "Настройте права доступа к защищенным ресурсам для {user}."
        ],
        'консультация': [
            "Как правильно настроить шифрование на съемном носителе?"
        ]
    },
    
    # 8. ВИДЕОНАБЛЮДЕНИЕ
    'Система видеонаблюдения': {
        'не работает': [
            "Не работает камера {camera_id} в {location}. Черный экран.",
            "Видеорегистратор в {location} завис, нет записи.",
            "Плохое качество изображения с камеры {camera_id}."
        ],
        'настройка': [
            "Настройте запись с камеры {camera_id} на более длительный срок.",
            "Добавьте новую камеру в {location} в систему видеонаблюдения.",
            "Настройте удаленный доступ к камерам для службы безопасности."
        ]
    },
    
    # 9. ВКС И ТЕЛЕФОНИЯ
    'Видео-конференц-связь': {
        'не работает': [
            "Не работает ВКС оборудование в переговорной {location}.",
            "Плохо слышно собеседника при ВКС с {counterparty}.",
            "Нет изображения на экране в переговорной {location}."
        ],
        'настройка': [
            "Настройте ВКС для совещания с {counterparty} в {datetime}.",
            "Подключите и настройте новое ВКС оборудование в {location}."
        ]
    },
    'IP-телефония': {
        'не работает': [
            "Не работает телефон у пользователя {user}. Нет гудка.",
            "Плохая слышимость при разговоре по телефону {number}.",
            "Телефон {number} не регистрируется в сети."
        ],
        'настройка': [
            "Настройте переадресацию с телефона {number} на {other_number}.",
            "Подключите новый телефон для пользователя {user}."
        ]
    },
    
    # 10. ДОСТУП И УЧЕТНЫЕ ЗАПИСИ
    'Учетные записи': {
        'доступ': [
            "Создайте учетную запись для нового сотрудника {user}.",
            "Заблокируйте учетную запись уволенного сотрудника {user}.",
            "Разблокируйте учетную запись пользователя {user}."
        ],
        'проблема': [
            "Не могу войти в систему под учетной записью {user}.",
            "Истек срок действия пароля у пользователя {user}.",
            "Учетная запись {user} заблокирована после нескольких неудачных попыток."
        ]
    },
    'Доступ к файловым ресурсам': {
        'доступ': [
            "Откройте доступ к папке {folder} для пользователя {user}.",
            "Нет прав на запись в папку {folder} у пользователя {user}.",
            "Требуется доступ к файловому ресурсу {resource} для выполнения {task}."
        ]
    }
}

# --- 4.2 Словари для подстановки значений ---
locations_pool = [
    "Административный корпус, каб. 301", "Главный корпус, зал управления", "Учебный центр, ауд. 12",
    "Лабораторный корпус, ком. 45", "Промплощадка, бытовка №3", "Здание управления, этаж 2",
    "Турбинный цех", "Хранилище, пост охраны", "АБК, каб. 115", "Проходная №2"
]

software_pool = {
    'Windows': ['Windows 10', 'Windows 11', 'Windows Server 2019'],
    'Office': ['Microsoft Office', 'LibreOffice', 'Офисный пакет'],
    'Special': ['САПР', '1С:Предприятие', 'СЭД "Дело"', 'АСУ ТП', 'SCADA-систему'],
    'Security': ['Kaspersky', 'Dr.Web', 'Секрет']
}

functions_pool = [
    "печать документов", "сохранение файлов", "отправка почты", "авторизация",
    "подключение к базе", "формирование отчетов", "экспорт данных"
]

problems_pool = [
    "зависает при запуске", "выдает ошибку 0x80070005", "тормозит",
    "не видит принтер", "отключается от сети", "греется и шумит"
]

errors_pool = [
    "Timeout Error", "Access Denied", "Connection refused", 
    "DNS error", "Service unavailable"
]

server_names = [
    "SRV-DC01", "SRV-FS02", "SRV-SQL03", "SRV-APP04", "SRV-BACKUP05"
]

# --- 4.3 Функция для определения категории услуги ---
def get_service_category(service_name):
    """Определяет категорию услуги на основе ее названия"""
    service_lower = str(service_name).lower() if service_name else ""
    
    if 'рабочая станция' in service_lower or 'арм оператора' in service_lower:
        return 'Рабочая станция'
    elif 'арм руководителя' in service_lower or 'руководител' in service_lower:
        return 'АРМ руководителя'
    elif 'тонкий клиент' in service_lower or 'vdi' in service_lower:
        return 'Рабочая станция'
    elif 'сетевое' in service_lower or 'кабельн' in service_lower or 'лвс' in service_lower:
        return 'Сетевое оборудование'
    elif 'сервер' in service_lower:
        return 'Серверная инфраструктура'
    elif 'оргтехник' in service_lower or 'принтер' in service_lower or 'мфу' in service_lower:
        return 'Оргтехника'
    elif 'почт' in service_lower:
        return 'Корпоративная почта'
    elif 'безопасн' in service_lower or 'сзи' in service_lower or 'антивирус' in service_lower:
        return 'Средства защиты информации'
    elif 'видеонаблюд' in service_lower:
        return 'Система видеонаблюдения'
    elif 'вкс' in service_lower or 'конференц' in service_lower:
        return 'Видео-конференц-связь'
    elif 'телефон' in service_lower or 'ip-телефония' in service_lower:
        return 'IP-телефония'
    elif 'учетн' in service_lower or 'домен' in service_lower:
        return 'Учетные записи'
    elif 'доступ' in service_lower and 'файлов' in service_lower:
        return 'Доступ к файловым ресурсам'
    elif 'документооборот' in service_lower or 'сэд' in service_lower:
        return 'Доступ к файловым ресурсам'
    else:
        return 'Рабочая станция'  # Дефолтная категория

# --- 4.4 Функция определения типа запроса ---
def get_request_type_category(request_type):
    """Определяет категорию запроса для выбора шаблона"""
    req_lower = str(request_type).lower() if request_type else ""
    
    if 'установ' in req_lower:
        return 'установка'
    elif 'настрой' in req_lower or 'перенастр' in req_lower or 'конфиг' in req_lower:
        return 'настройка'
    elif 'обновл' in req_lower:
        return 'обновление'
    elif 'доступ' in req_lower:
        return 'доступ'
    elif 'консульт' in req_lower or 'подскаж' in req_lower or 'помощ' in req_lower:
        return 'консультация'
    elif 'заправ' in req_lower or 'картридж' in req_lower:
        return 'заправка'
    elif 'не работ' in req_lower or 'ошибк' in req_lower or 'сбой' in req_lower:
        return 'не работает'
    elif 'проблем' in req_lower:
        return 'проблема'
    else:
        return 'не работает'  # Дефолтный тип

# --- 4.5 Функция генерации описания ---
def generate_smart_description(service, request_type, user_id, location):
    """
    Генерирует осмысленное описание на основе категории услуги и типа запроса
    """
    import random
    
    # Определяем категории
    category = get_service_category(service)
    req_category = get_request_type_category(request_type)
    
    # Получаем шаблоны для категории
    category_templates = service_category_templates.get(category, service_category_templates['Рабочая станция'])
    
    # Получаем шаблоны для типа запроса
    templates = category_templates.get(req_category, category_templates.get('не работает', []))
    
    if not templates:
        templates = ["Требуется обслуживание {service} для пользователя {user} в {location}."]
    
    # Выбираем случайный шаблон
    template = random.choice(templates)
    
    # Подготавливаем переменные для подстановки
    vars_dict = {
        'user': user_id if pd.notna(user_id) else 'User_Unknown',
        'location': location if pd.notna(location) else random.choice(locations_pool),
        'software': random.choice(
            software_pool['Windows'] + 
            software_pool['Office'] + 
            software_pool['Special']
        ),
        'function': random.choice(functions_pool),
        'problem': random.choice(problems_pool),
        'error': random.choice(errors_pool),
        'equipment': random.choice(['коммутатор', 'маршрутизатор', 'медиаконвертер', 'точка доступа']),
        'server_name': random.choice(server_names),
        'service': service,
        'resource': random.choice(['сетевую папку', 'базу данных', 'корпоративный портал', '1С']),
        'task': random.choice(['работы с документами', 'отправки отчетности', 'участия в ВКС']),
        'camera_id': f"IPCAM-{random.randint(101, 199)}",
        'number': f"{random.randint(100, 999)}-{random.randint(10, 99)}",
        'other_user': f"User_{hashlib.md5(str(random.randint(100, 200)).encode()).hexdigest()[:8]}",
        'counterparty': random.choice(['Москва', 'Санкт-Петербург', 'Смоленск', 'Курск']),
        'datetime': f"{random.randint(10, 30)}.03.2026 в {random.randint(9, 18)}:00",
        'folder': random.choice(['Отчеты_2026', 'Общая', 'Сметы', 'Документация', 'Проекты'])
    }
    
    # Добавляем вежливость
    if random.random() > 0.7:
        vars_dict['greeting'] = 'Добрый день'
        template = "{greeting}! " + template
    
    try:
        description = template.format(**vars_dict)
    except KeyError as e:
        # Если какой-то ключ не найден, используем простой шаблон
        description = f"Требуется обслуживание {service} для пользователя {user_id}."
    
    return description

# --- 4.6 Применяем генерацию ко всем строкам ---
print("\nГенерируем новые описания для всех заявок...")

# Определяем колонки
description_col = None
service_col = None
request_col = None
user_col = None
location_col = None

for col in applications_anon.columns:
    if 'Описание' in col:
        description_col = col
    elif 'Услуга' in col:
        service_col = col
    elif 'Вид запроса' in col:
        request_col = col
    elif 'Получатель услуг' in col or 'Инициатор' in col:
        user_col = col  # Берем первый попавшийся пользователь
    elif 'Расположение' in col:
        location_col = col

# Проверяем наличие колонок
if not all([description_col, service_col, request_col]):
    print(f"  description_col: {description_col}")
    print(f"  service_col: {service_col}")
    print(f"  request_col: {request_col}")
    raise ValueError("Не найдены все необходимые колонки!")

# Если нет колонки с пользователем, создаем заглушку
if not user_col:
    applications_anon['temp_user'] = 'User_Unknown'
    user_col = 'temp_user'

if not location_col:
    applications_anon['temp_location'] = 'Не указано'
    location_col = 'temp_location'

# Сохраняем оригинальные описания
original_descriptions = applications_anon[description_col].copy()

# Генерируем новые
new_descriptions = []
for idx, row in applications_anon.iterrows():
    new_desc = generate_smart_description(
        service=row[service_col],
        request_type=row[request_col],
        user_id=row[user_col],
        location=row[location_col]
    )
    new_descriptions.append(new_desc)

applications_anon[description_col] = new_descriptions

# Удаляем временные колонки если создавали
if 'temp_user' in applications_anon.columns:
    applications_anon.drop('temp_user', axis=1, inplace=True)
if 'temp_location' in applications_anon.columns:
    applications_anon.drop('temp_location', axis=1, inplace=True)

# --- 4.7 Проверка ---
print(f"\nПроверка:")
print(f"  Всего строк обработано: {len(applications_anon)}")
print(f"\nПримеры ДО замены (первые 3):")
for i in range(min(3, len(original_descriptions))):
    print(f"  {i+1}. {str(original_descriptions.iloc[i])[:100]}...")

print(f"\nПримеры ПОСЛЕ замены (первые 3):")
for i in range(min(3, len(new_descriptions))):
    print(f"  {i+1}. {new_descriptions[i][:150]}")

# --- 4.8 Проверка соответствия категориям ---
print(f"\nПроверка соответствия услуге и виду запроса:")
sample_indices = random.sample(range(len(applications_anon)), min(8, len(applications_anon)))
for idx in sample_indices:
    service = applications_anon.iloc[idx][service_col]
    req_type = applications_anon.iloc[idx][request_col]
    desc = applications_anon.iloc[idx][description_col]
    category = get_service_category(service)
    print(f"\n  [{category}]")
    print(f"  Услуга: {service}")
    print(f"  Вид запроса: {req_type}")
    print(f"  Описание: {desc}")
```
