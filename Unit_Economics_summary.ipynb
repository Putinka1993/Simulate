## Unit Economics summary

# 📊 Юнит-экономика продукта

Этот конспект объединяет два ключевых блока:
1. **Финансовые метрики продукта с учётом LT (Lifetime)**
2. **Маркетинговое привлечение и эффективность рекламных кампаний**

---

## 💰 Финансовые показатели с учетом LT

| Метрика | Полное название | Что показывает | Формула | Комментарий |
|----------|-----------------|----------------|----------|--------------|
| **LT, мес** | Lifetime | Среднее время жизни клиента в продукте | `1 / Churn Rate` или `avg(last_date - first_date)` | Показывает, сколько в среднем клиент активен |
| **ARPPU, руб** | Average Revenue Per Paying User | Средний доход на одного платящего пользователя | `Revenue / Paying Users` | Используется для расчёта LTV |
| **LTV, руб** | Lifetime Value | Доход, который клиент приносит за всё время жизни | `ARPPU × LT` | Ключевой показатель ценности клиента |
| **Revenue, руб** | Выручка | Все поступления без вычета расходов | `Σ(доходов клиентов)` | Полная сумма продаж |
| **Profit, руб** | Прибыль | Чистая прибыль после расходов | `Revenue − Costs` | Показывает рентабельность бизнеса |
| **Profit per client, руб** | Прибыль на клиента | Средняя прибыль на одного клиента | `Profit / Users` | Отражает эффективность работы с клиентом |
| **Margin, %** | Маржинальность | Доля прибыли в выручке | `(Profit / Revenue) × 100%` | |
| **LTV / CAC** | Lifetime Value to CAC | Отношение дохода от клиента к стоимости привлечения | `LTV / CAC` | >1 — прибыльно, <1 — убыточно |
| **ROMI (ROI), %** | Return on Marketing Investment | Доходность маркетинга | `((Revenue − Marketing Costs) / Marketing Costs) × 100%` | Показывает эффективность инвестиций |

---

## 📢 Маркетинговое привлечение

| Метрика | Полное название | Что показывает | Формула / Как считается | Комментарий |
|----------|-----------------|----------------|--------------------------|--------------|
| **Spent, руб** | Расходы на рекламу | Сколько денег потрачено на кампанию | — | Основной бюджет кампании |
| **Impressions, шт** | Показы | Сколько раз реклама показалась пользователям | — | Кол-во всех показов |
| **Reach, шт** | Охват | Сколько уникальных пользователей увидели рекламу | — | Один человек мог увидеть рекламу несколько раз |
| **Avg impressions, шт** | Среднее число показов на пользователя | `Impressions / Reach` | Помогает оценить частоту контакта |
| **CPM, руб** | Cost per Mille | Стоимость 1000 показов | `(Spent / Impressions) × 1000` | Базовая метрика закупки трафика |
| **Clicks, шт** | Клики | Количество кликов по рекламе | — | Следующий этап после показов |
| **CTR, %** | Click Through Rate | Эффективность креатива (сколько кликнули) | `(Clicks / Impressions) × 100%` | Отражает привлекательность рекламы |
| **CPC, руб** | Cost per Click | Стоимость клика | `Spent / Clicks` | Чем ниже, тем лучше |
| **Bounce rate, %** | Показатель отказов | Пользователи, ушедшие сразу после клика | `(Bounced / Clicks) × 100%` | Сигнал, насколько лендинг релевантен |
| **Leads, шт** | Лиды | Количество регистраций или заявок | — | Потенциальные клиенты |
| **CR1, %** | Conversion Rate 1 | Конверсия из кликов в лиды | `(Leads / Clicks) × 100%` | Эффективность лендинга |
| **CPL, руб** | Cost per Lead | Стоимость одного лида | `Spent / Leads` | Показывает цену заявки |
| **vLeads, шт** | Valid Leads | Кол-во валидных (качественных) лидов | — | После фильтрации спама |
| **vCR, %** | Validation Conversion Rate | Доля валидных лидов | `(vLeads / Leads) × 100%` | Показывает качество трафика |
| **vCPL, руб** | Cost per Valid Lead | Стоимость валидного лида | `Spent / vLeads` | Реальная цена хорошей заявки |
| **Purchases, шт** | Покупки | Количество оплат | — | Финальный этап воронки |
| **CR2, %** | Conversion Rate 2 | Конверсия из лидов в покупки | `(Purchases / Leads) × 100%` | Эффективность отдела продаж |
| **CAC, руб** | Customer Acquisition Cost | Стоимость привлечения одного клиента | `Spent / Purchases` | Ключевой показатель окупаемости маркетинга |

---

## ⚙️ Связь между блоками

| Финансы | Маркетинг | Связь |
|----------|------------|-------|
| **LTV** | **CAC** | Сравнение ценности клиента и стоимости привлечения |
| **Revenue** | **Spent** | ROMI показывает возврат инвестиций |
| **Profit per client** | **CPC / CPL** | Определяет, насколько дорого обходится продажа |
| **Margin** | **CR1 / CR2** | Зависит от конверсии по воронке и себестоимости |

---

## 💡 Основная идея

> Юнит-экономика показывает, **окупается ли привлечение пользователей**,  
> и **насколько эффективен весь путь клиента — от клика до покупки**.

---

## 🧠 Мини-гайд: Как использовать

1. **Для анализа продукта:**
   - Определи `LTV`, `ARPPU`, `LT` — это то, сколько денег приносит клиент.
   - Рассчитай `CAC` — сколько стоит его привлечение.
   - Сравни `LTV / CAC`:  
     - 🔵 > 1 → бизнес окупается,  
     - 🔴 < 1 → теряешь деньги.

2. **Для маркетинга:**
   - Построй воронку по шагам: `Impressions → Clicks → Leads → Purchases`.
   - На каждом этапе оцени конверсии (`CTR`, `CR1`, `CR2`).
   - Проверь, где теряешь пользователей (например, высокий bounce rate).

3. **Для принятия решений:**
   - Определи, где “дырит” воронка.
   - Считай ROMI → какие каналы реально приносят прибыль.
   - Прогнозируй окупаемость по `LTV`, если увеличишь `LT` или `ARPPU`.

---

## 📘 Пример структуры в pandas

```python
# расчет ключевых метрик
df['CPM'] = (df['Spent'] / df['Impressions']) * 1000
df['CTR'] = (df['Clicks'] / df['Impressions']) * 100
df['CPC'] = df['Spent'] / df['Clicks']
df['CR1'] = (df['Leads'] / df['Clicks']) * 100
df['CPL'] = df['Spent'] / df['Leads']
df['CR2'] = (df['Purchases'] / df['Leads']) * 100
df['CAC'] = df['Spent'] / df['Purchases']


🧩 Вывод

Этот шаблон — универсальный инструмент анализа юнит-экономики.

Его цель — понять:

💸 сколько стоит клиент,

📈 сколько он приносит,

🧭 и на каком этапе ты теряешь прибыль.

Работает в Pandas, SQL, Excel, Power BI, Tableau — везде, где можно посчитать конверсии и средние значения.

### 💬 Как и зачем ты это используешь

Ты применяешь этот шаблон, когда хочешь:

1. **Понять эффективность бизнеса.**  
   → Показывает, окупается ли каждый привлечённый пользователь.

2. **Увидеть слабые места в воронке.**  
   → Например, CTR высокий, но CR2 низкий — значит, плохой лендинг или продукт.

3. **Оценить каналы маркетинга.**  
   → Сравнивай CAC и ROMI по источникам: Google Ads, VK, TikTok и т.д.

4. **Построить финансовую модель.**  
   → Зная LTV и CAC, можно прогнозировать прибыль, бюджет, сроки окупаемости.

5. **Обосновать решения для инвесторов / руководства.**  
   → LTV/CAC > 3 — сигнал, что модель устойчива и масштабируема.
