# 3. Стратифицированная выборка

```python
# Чтобы сделать стратифицированную выборку, сгруппируем набор данных по столбцу 'col_name' и выберем случайным образом n строк из каждой группы. 

# При такой выборке мы не можем гарантировать, что в группе будет n уникальных строк, поэтому либо нужно добавить параметр replace = True, 
# тогда выборка будет повторная, либо заранее оценить самую малочисленную группу и поставить n равный или меньше найденного значения.
```

```python
stratified_sample = CODE_RUN.groupby('problem_id', group_keys=False).apply(lambda x: x.sample(n = 3, replace = True))
stratified_sample.head(5)
```

```python
# При таком выборе мы добавляем в выборку равное количество экземпляров из каждой группы, что не совсем правильно.

# Предположим, мы оцениваем среднее число покупок в магазине косметики и решили сделать стратифицированную выборку. 
#В базе клиентов 90 девушек и 10 мужчин. Когда мы будем делать выборку приведенным выше способом, например, поставим n=10, то получим 10 мужчин и 10 женщин.
# Женщины покупают косметику с завидной регулярностью, а вот мужчины от силы 3 раза в год: день рождения жены, Новый год и 8 марта.
# В итоге мы получим выборку с равной пропорцией элементов из каждой группы, хотя в ГС мы видим, что одна группа в 9 раз больше другой. 
# И если по данной выборке мы будем делать заключение о всей ГС, то получим искаженный результат.

# Что же делать?

# Мы можем добавить параметр frac. Тогда из каждой группы мы возьмем равную часть от каждой группы. Например, frac=0.5 – это 50% от группы.
```

```python
stratified_sample = CODE_RUN.groupby('problem_id', group_keys=False).apply(lambda x: x.sample(frac=0.5))
stratified_sample.head(5)
```

```python
# Теперь у нас в выборке 45 женщин и 5 мужчин. С этим уже можно работать.

# Но что если мы хотим указать размер выборки, и при этом чтобы доля представителей из каждой группы соответствовала доле этой группы в генеральной совокупности? 
# Например, мы хотим, чтобы выборка состояла из 70 клиентов, среди которых 90% женщин и 10% мужчин (как раз как в ГС).

# Тогда нам нужно указать размер выборки N и воспользоваться следующей конструкцией:
```

```python
# выбираем 1000 элементов из датафрейма df
N = 1000
df = CODE_RUN
stratified_proportional = df.groupby('problem_id', group_keys=False).apply(lambda x: x.sample(int(np.rint(N*len(x)/len(df)))))
stratified_proportional.head(5)
```

```python
# Количество, которое нужно взять из каждой группы, высчитывается следующим образом: 
# объем группы делим на объем всех данных – это доля, которую занимает эта группа в генеральной совокупности. 
#Умножаем долю на требуемый размер выборки N и получаем количество данных, которые нужно включить в выборку из конкретной группы.

# Теперь пару слов о конструкции округления. Сначала округляем до ближайшего целого с помощью np.rint, затем преобразовываем в тип int.
```

# 4. Кластерная и серийная выборка

```python
# Теперь мы хотим сделать кластерную выборку пользователей из файла 'itresume-users.csv', например, по периоду регистрации пользователей (по кварталу).

#Для начала считываем файл:
```

```python
USERS = pd.read_csv(r'C:\Users\Incognitus\Documents\git_hub\Simulate\TRASH\itresume-users.csv', sep = ',')
USERS.head(5)
```

```python
# Создадим столбец с кварталом: 
```

```python
USERS['period'] = pd.to_datetime(USERS['date_joined']).dt.to_period('Q').apply(lambda x: x.strftime('%F-Q%q'))
```

```python
# Посмотрим, как распределились наши данные по кварталам: сгруппируем таблицу USERS по 'period' и посчитаем количество строк:
```

```python
USERS.groupby('period')['id'].count()
```

```python
# Получили четыре периода.

# У нас два периода малочисленные, поэтому можно было бы в жизни их вообще не рассматривать, но для примера оставим их.

# Теперь нам нужно случайным образом выбрать, например, два периода, из которых будем случайно выбирать элементы:
```

```python
import random
clusters = random.sample(USERS['period'].unique().tolist(), 2)
```

```python
# Здесь USERS['period'].unique() оставит уникальные значения периодов, .tolist() преобразует их в список, 2 - параметр, который показывает, что нужно выбрать 2 случайных элемента. 

# Теперь выбираем всех пользователей из выбранных кластеров. 
```

```python
serial_sample = USERS[USERS['period'].isin(clusters)]
```

```python
# Мы получили серийную выборку – взяли все экземпляры из каждого случайно выбранного кластера.

# Чтобы получить кластерную выборку, применим к серийной выборке формулу для стратифицированной выборки, например, с параметром frac:
```

```python
cluster_sample = serial_sample.groupby('period', group_keys=False).apply(lambda x: x.sample(frac=0.7))
```

```python
# Мы разобрали все типы вероятностных выборок. 
# На практике невероятностные выборки нас мало интересуют. Чаще всего они не являются репрезентативными.
# Их можно использовать в корыстных целях, например, для подтасовки фактов или получения заведомо нужной информации. 
```

## Часть 1. Кластеризованная выборка из 20 пользователей

**Что требуется:** взять 20 пользователей из таблицы USERS так, чтобы они представляли 3 случайные компании в тех же пропорциях, что и в исходной таблице.

**Алгоритм действий:**

1. **Найти пропорции компаний в генеральной совокупности**  
   Посчитать, сколько всего пользователей в каждой компании, и определить долю каждой компании от общего числа пользователей.

2. **Выбрать 3 случайные компании**  
   Из всех существующих company_id случайным образом выбрать 3.

3. **Определить, сколько пользователей взять из каждой компании**  
   Умножить долю каждой из 3 выбранных компаний на 20 (нужный размер выборки) и округлить до целого числа. Сумма должна равняться 20 (если не хватает — добавить недостающих в одну из компаний).

4. **Из каждой компании случайно выбрать нужное количество пользователей**  
   Для каждой компании отобрать случайные user_id в количестве, рассчитанном на шаге 3.

**Итог:** 20 user_id, пропорционально представляющих 3 случайные компании.
